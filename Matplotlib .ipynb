# MatplotLib Tutorial

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

    Generally easy to get started for simple plots
    Support for custom labels and texts
    Great control of every element in a figure
    High-quality output in many formats
    Very customizable in general



```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples

x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##plotting using matplotlib 

##plt scatter : scatter use to just scattter the datasets

plt.scatter(x,y,c='r')
plt.xlabel('X axis') #use to provide name of x axis
plt.ylabel('Y axis')
plt.title('Graph in 2D') #use to give the title of whole graph
plt.savefig('Test.png')  #use to save the graph
```

```python
y= x*x
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots

plt.subplot(2,2,1) #(no rows,no col,pos) 
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
print(np.pi)

# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show()

```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## Bar plot

x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()

```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=40)
plt.title("histogram") 
plt.show()

#bins is just use to tell the density oor count of the variables
```

# BOX PLOT USING MATPLOT

```python
#box plot generally used to plot percentile 
data = [np.random.normal(0, std, 100) for std in range(1, 4)]   #3 array creating using for loop

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);   #small circle tells the outliers of data sets

#vert  used to plot box in vertical and patch_artist use to color the box
```

# pie chart

```python
# Data to plot
labels = 'C++', 'Ruby', 'Java' ,'Python'
sizes = [215, 130, 245, 210]  # these size allotes with the percentage of whole

colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0, 0.4, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
